```python
#checking outliers is there or not
# but no outliers found from the output
from scipy import stats

# Columns to test
columns_to_test = [
    'Users', 'Sessions', 'Engaged sessions_x', 'Average engagement time per session',
    'Engaged sessions per user_x', 'Events per session', 'Engagement rate_x',
    'Event count_x', 'Conversions_x', 'Total revenue_x', 'New users',
    'Engaged sessions_y', 'Engagement rate_y', 'Engaged sessions per user_y',
    'Average engagement time', 'Event count_y', 'Conversions_y', 'Total revenue_y'
]

alpha = 0.05  # Significance level

for column in columns_to_test:
    # Perform t-test
    t_stat, p_value = stats.ttest_1samp(df_merge[column], df_merge[column].mean())

    # Interpret the results
    if p_value < alpha:
        print(f"Reject null hypothesis for column '{column}': There is a statistically significant difference.")
    else:
        print(f"Fail to reject null hypothesis for column '{column}': There is no statistically significant difference.")
```

```python
import matplotlib.pyplot as plt

# Columns to test
columns_to_test = [
    'Users', 'Sessions', 'Engaged sessions_x', 'Average engagement time per session',
    'Engaged sessions per user_x', 'Events per session', 'Engagement rate_x',
    'Event count_x', 'Conversions_x', 'Total revenue_x', 'New users',
    'Engaged sessions_y', 'Engagement rate_y', 'Engaged sessions per user_y',
    'Average engagement time', 'Event count_y', 'Conversions_y', 'Total revenue_y'
]

# Calculate p-values for each column
p_values = []
for column in columns_to_test:
    t_stat, p_value = stats.ttest_1samp(df_merge[column], df_merge[column].mean())
    p_values.append(p_value)

# Plot the p-values
plt.figure(figsize=(12, 6))
plt.bar(columns_to_test, p_values, color='skyblue')
plt.axhline(y=alpha, color='red', linestyle='--', label='Significance Level (alpha)')
plt.xticks(rotation=45)
plt.ylabel('p-value')
plt.title('p-values from t-test for Each Column')
plt.legend()
plt.tight_layout()
plt.show()
```
